**Exercício da variabilidade**

Você é o gerente de um time de futebol e precisa escolher entre dois atacantes baseado na quantidade de gols feitos em cinco temporadas. Justifique a sua escolha utilizando os medidores de dispersão.

Gols realizados nas temporadas 14/15, 15/16, 16/17, 17/18 e 18/19:

Jogador 1 — 32, 29, 33, 30, 32

Jodador 2 — 39, 31, 20, 28, 23



```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Dados dos jogadores
j1 = [32, 29, 33, 30, 32]
j2 = [39, 31, 20, 28, 23]
```

```python
avg1 = np.mean(j1)
avg2 = np.mean(j2)

print(avg1)
print(avg2)
```

**A média de gols nas cinco temporadas nao e muito diferente, portanto, seria possível escolher entre um dos dois, se você precisar de um jogador que marque gols consistentemente nas próximas temporadas, ou seja, nao apenas todos os gols em uma temporada, é preciso analisar nao apenas a média.**


```python
plt.figure(figsize = (15, 6))
seasons = ['14/15', '15/16', '16/17', '17/18', '18/19']
plt.plot(seasons, j1, marker='o', color='#003366', label='Jogador 1')
plt.axhline(y = avg1, color='#6684a3', linestyle='--')
plt.plot(seasons, j2, marker='o', color='#be4038', label='Jogador 2')
plt.axhline(y = avg2, color='#d88c87', linestyle='--')
plt.xlabel('Temporadas')
plt.ylabel('Gols')
plt.title('Comparacao')
plt.legend()# This will pop up the graph
plt.show()
```

```python
var1 = np.var(j1)
var2 = np.var(j2)
print(f'Variancia do jogador 1 em cinco temporadas: {round(var1, 2)}')
print(f'Variancia do jogador 2 em cinco temporadas: {round(var2, 2)}')# Expected Output:
```

```python
std1 = np.std(j1)
std2 = np.std(j2)
print(f'Desvio padrao do jogador 1 em cinco temporadas: {round(std1, 2)}')
print(f'Desvio padrao do jogador 2 em cinco temporadas: {round(std2, 2)}')
```

**Nesse caso nao e interessante ter um jogador que possa ter altos e baixos. O jogador 1 marcará cerca de 31 gols em cada temporada.**
