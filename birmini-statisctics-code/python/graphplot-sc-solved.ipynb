```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Criando dataset para facilitar o ordenação dos elementos
df = pd.DataFrame(
    {
        'prejuizo' : [3.5, 0.1, 0.6, 2.4, 0.1, 33, 36.75, 23.7, 0.1, 1.75] 
    }
)

df.index = ['Moldagem solta', 'Solda quebrada', 'Centro de moldagem deslocado',
            'Lateral da moldagem deslocado', 'Moldagem arranhada', 'Moldagem dentada',
            'Plastico arranhado', 'Limpeza incompleta', 'Orificio deslocado', 'Pino deslocado'
]


# # Ordenando valores de forma descendente
df = df.sort_values(by = 'prejuizo', ascending = False)


# Acumulando a percentagem relativa
df['cumperc'] = df['prejuizo'].cumsum()/df['prejuizo'].sum() * 100

# Criando figura e gráfico
fig, ax = plt.subplots()

# Criação do gráfico de barras
ax.bar(df.index, df['prejuizo'])

# Definindo segundo
ax2 = ax.twinx()

# Criando linha
ax2.plot(df.index, df['cumperc'], marker = "D", color = "red")

# Rotacionando o eixo x
ax.set_xlabel('Tipo de defeito')
ax.set_ylabel('Prejuizo total')
ax2.set_ylabel("Percentagem")
ax.tick_params(axis='x', labelrotation = 45)

fig.set_size_inches((22, 10.5))
plt.show()
```
